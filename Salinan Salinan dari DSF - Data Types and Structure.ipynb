# **1. Import Library**

# **2. Read Dataset**

```python
import pandas as pd
from sklearn import datasets

# Memuat dataset iris dari scikit-learn
iris = datasets.load_iris()

X = iris.data    # inputan untuk machine learning
y = iris.target  # output yang dinginkan dari machine learning

# Mengonversi data fitur dan target menjadi DataFrame
df_X = pd.DataFrame(X, columns=iris.feature_names)
df_y = pd.Series(y, name='target')
```

```python
# Gabungkan fitur dan target dalam satu DataFrame
df = pd.concat([df_X, df_y], axis=1)

df.head(20)
```

# **3. Split Data**

```python
from sklearn.model_selection import train_test_split

# Membagi data menjadi train dan test
X_train, X_test, y_train, y_test = train_test_split(df_X, df_y, test_size=0.3, random_state=42)
```

# **4. Train the Model**

```python
# tipe datanya itu termasuk supervised atau unsupervised
# klasifikasi atau regresi --> supervised
# clustering --> unsupervised

from sklearn.tree import DecisionTreeClassifier

# Membuat dan melatih model Decision Tree
model = DecisionTreeClassifier(random_state=42, criterion='log_loss', min_samples_split=5, min_samples_leaf=3)
model.fit(X_train, y_train)
```

# **5. Predict & Evaluate**

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

# 4. Memprediksi dan mengevaluasi
y_pred = model.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)

print("Laporan Klasifikasi:")
print(f"Akurasi: {accuracy * 100:.4f}%")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Confusion Matrix
cm = confusion_matrix(y_test, y_pred)

# Plot Confusion Matrix using seaborn
plt.figure(figsize=(6,5))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues",
            xticklabels=iris.target_names, yticklabels=iris.target_names)
plt.xlabel("Prediksi")
plt.ylabel("Aktual")
plt.title("Confusion Matrix")
plt.show()
```

```python
from sklearn.metrics import accuracy_score, classification_report

print(classification_report(y_test, y_pred, target_names=iris.target_names))
```
